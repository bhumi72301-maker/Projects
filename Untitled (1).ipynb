```python
from ultralytics import YOLO
import cv2
from IPython.display import display, clear_output, Image
import numpy as np

# Load YOLOv8 model (Nano version is fastest)
model = YOLO('yolov8n.pt')  # Make sure yolov8n.pt is in your directory or use full path

# Use built-in laptop camera
cap = cv2.VideoCapture(0)  # Keep 0 for default laptop camera

# Check if camera opened successfully
if not cap.isOpened():
    print(" Unable to access the camera. Try changing the index (e.g., 1 or 2).")
else:
    print(" Camera accessed successfully.")

while True:
    ret, frame = cap.read()
    if not ret or frame is None:
        print(" Failed to grab frame from camera.")
        break

    # Run YOLOv8 model
    results = model(frame, stream=True)

    vehicle_count = 0  # Initialize vehicle counter

    for r in results:
        for box in r.boxes:
            cls = int(box.cls[0])
            label = model.names[cls]
            if label in ['car', 'truck', 'bus', 'motorbike']:
                vehicle_count += 1
                x1, y1, x2, y2 = map(int, box.xyxy[0])
                conf = float(box.conf[0])
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(frame, f'{label} {conf:.2f}', (x1, y1 - 10),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    # Determine car density
    if vehicle_count <= 3:
        density = "Low"
        color = (0, 255, 0)
    elif vehicle_count <= 6:
        density = "Moderate"
        color = (0, 255, 255)
    else:
        density = "High"
        color = (0, 0, 255)

    # Show vehicle count and density on frame
    cv2.putText(frame, f'Vehicles: {vehicle_count} | Density: {density}',
                (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)

    # Resize and display frame
    resized = cv2.resize(frame, (640, 480))
    _, img_encoded = cv2.imencode('.jpg', resized)
    clear_output(wait=True)
    display(Image(data=img_encoded.tobytes()))

# Release resources
cap.release()
cv2.destroyAllWindows()
```
